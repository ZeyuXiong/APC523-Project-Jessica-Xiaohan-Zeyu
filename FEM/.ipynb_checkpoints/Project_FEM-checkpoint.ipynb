```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import numpy.linalg
import scipy.linalg as spla
# %matplotlib tk

import quadrilateral_element
from element_operations.poisson import *
from utils import *
```

```python
def infnorm(exact,approx):
    return(max(abs(exact-approx)))

def relerr(exact,approx):
    return abs((exact-approx)/exact)

def abserr(exact,approx):
    return abs((exact-approx))
```

```python
def uexact(x,x0,y0):
    N=100
    uexact = -sum([-4/(((2*j-1)**2+(2*k-1)**2)*np.pi**2)*np.sin((2*j-1)*np.pi*x0)*np.sin((2*k-1)*np.pi*y0)*np.sin((2*j-1)*np.pi*x[0])*np.sin((2*k-1)*np.pi*x[1]) for j in range(1,N) for k in range(1,N)])
    return uexact
```

```python
#Gauss 0.01
exponents=np.arange(1,8,1)
orders=[1]
Mat5=[np.zeros(16641),np.zeros(16641),np.zeros(16641),np.zeros(16641),np.zeros(16641),np.zeros(16641),np.zeros(16641)]
coords5=[np.zeros((16641,2)),np.zeros((16641,2)),np.zeros((16641,2)),np.zeros((16641,2)),np.zeros((16641,2)),np.zeros((16641,2)),np.zeros((16641,2))]

import numpy as np

for p in orders:
    print('Order = ',p)
    for ex in exponents:
        if __name__ == "__main__":

            # Define the order of the polynomial interpolant
            poly_order = p

            # Define the width of the domain
            w = 1.

            # Create the mesh of a square
            coordinates = np.array([ (0,0), (w,0), (w,w), (0,w) ]) 
            coordinates2 = np.array([ (0,0), (w,0), (w,w), (0,w) ]) 

            # Connectivity 
            connectivity = np.array([[0,1,2,3]])
            connectivity2 = np.array([[0,1,2,3]])

            # Refine a few times
            for i in range(ex):
                coordinates, connectivity = subdivide_mesh( coordinates, connectivity )
            num_elements=4**(ex)
            N=2**(ex)
            
            #For plotting     
            for i in range(int(poly_order/2+ex)):
                coordinates2, connectivity2 = subdivide_mesh( coordinates2, connectivity2 )

            # Construct an array of element obejects 
            elements = []

            for e in range(len( connectivity ) ):
                elements.append( quadrilateral_element.element( e,\
                        coordinates, connectivity, poly_order) )

            # Create an instance of the local to global map 
            # for the current mesh
            l2g = quadrilateral_element.local_to_global_map( poly_order, connectivity )

            # Print out 
            total_dofs =  l2g.get_total_dof()
            print( 'Total dofs : %i ' %( total_dofs  ))

            # Initialize stiffness matrix and force vector
            K = np.zeros( (total_dofs, total_dofs ) )
            F = np.zeros( total_dofs )

            # Initialize the element operation 
            x0,y0= 0.5,0.5
            L= 0.01

            #Gaussian/Normal
            f = lambda x: (1/(2*np.pi*L**2))*np.exp(-((x[0]-x0)**2+(x[1]-y0)**2)/(2*L**2))

            poisson_problem = poisson( f )  

            print('Assemble system of equations')

            # Assemble global systems
            assemble_global(l2g, poisson_problem , elements, K, F )

            print('System of equations asembled')

            # Find all degrees of freedom on boundary
            boundary_dofs = l2g.get_boundary_dofs() 

            # Get the values of the boundary conditions
            boundary_vals = boundary_dofs*0 

            # Apply boundary conditions
            apply_bc(boundary_dofs, boundary_vals, K, F )

#             plt.matshow(K)

            # Solve
            u = spla.solve( K, F ) 
            Mat5[ex-1]=u
            coords5[ex-1]=coordinates2

            # Plot approx
            plot_contour_3d( coordinates2, u )
```

```python
#0.01
error5=np.zeros(7)
error5FEM=np.zeros(7)

for n in range(7):
    if n == 0:
        pass
    else:
        coordinates5=list(coords5[n])
        for i in range(len(coordinates5)):
            coordinates5[i]=tuple(coordinates5[i])
        d5=dict( zip( coordinates5, Mat5[n]))
        nodes=np.sqrt(len(coords5[n]))
        idxs5=np.linspace(0,1,nodes)
        y5=np.zeros(len(idxs5))
        for idx,i in enumerate(idxs5):
            y5[idx]=d5[(i,0.25)]
        plt.plot(idxs5, y5,'-')

        plt.xlabel('x')
        plt.ylabel('y')
        plt.grid(True)
        plt.show()

        integral5=np.trapz(y5, x=idxs5)
        print(integral5)
        error5[n]=integral-integral5
        error5FEM[n]=abs(0.06641244842207733-integral5)
print(error5FEM)
```
